# Visualization of MNIST and Fashion MNIST Latent Factors

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load MNIST dataset
mnist = fetch_openml("mnist_784", version=1)
fashion_mnist = fetch_openml("Fashion-MNIST", version=1)

# Number of samples in each dataset
num_smp = 1000
X_mnist, y_mnist = mnist.data[:num_smp], mnist.target[:num_smp]
X_fashion_mnist, y_fashion_mnist = fashion_mnist.data[:num_smp], fashion_mnist.target[:num_smp]
```

```python
# Import Kernel PCA
from sklearn.decomposition import PCA

# Apply PCA to MNIST and Fashion MNIST
n_components = 2  # Choose the number of components
model = PCA(n_components=n_components)
X_mnist_reduced = model.fit_transform(X_mnist)
X_fashion_mnist_reduced = model.fit_transform(X_fashion_mnist)
```

```python
# Visualize the first two principal components
# set the size of the plot
plt.figure(figsize=(6, 5))
plt.grid(alpha=0.5)
plt.scatter(X_mnist_reduced[:, 0], X_mnist_reduced[:, 1], c='royalblue', s=60, alpha=0.5, label="MNIST", edgecolors='k')
plt.scatter(X_fashion_mnist_reduced[:, 0], X_fashion_mnist_reduced[:, 1], c='darkorange', s=60, alpha=0.5, label="Fashion MNIST", edgecolors='k')
plt.xlabel("Latent $Z_1$")
plt.ylabel("Latent $Z_2$")
plt.legend()

# Save the figure
plt.savefig("images/mnist_latentZ.pdf", dpi=300)
```

```python
# Import Kernel PCA
from sklearn.decomposition import TruncatedSVD

# Apply PCA to MNIST and Fashion MNIST
n_components = 2  # Choose the number of components
model = TruncatedSVD(n_components=n_components)
X_mnist_reduced = model.fit_transform(X_mnist)
X_fashion_mnist_reduced = model.fit_transform(X_fashion_mnist)
```

```python
# Visualize the first two principal components
plt.figure(figsize=(6, 5))
plt.grid(alpha=0.5)
plt.scatter(X_mnist_reduced[:, 0], X_mnist_reduced[:, 1], c='royalblue', edgecolor="k", s=60, alpha=0.5, label="MNIST")
plt.scatter(X_fashion_mnist_reduced[:, 0], X_fashion_mnist_reduced[:, 1], c='darkorange', edgecolor="k", s=60, alpha=0.5, label="Fashion MNIST")
plt.xlabel("Latent $Z_1$", fontsize=12)
plt.ylabel("Latent $Z_2$", fontsize=12)
plt.legend()
plt.tight_layout()

# Save the figure
plt.savefig("images/mnist_latentZ.pdf", dpi=300)
```
